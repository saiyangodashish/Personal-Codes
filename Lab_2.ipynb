```python
import pandas as pd
import matplotlib as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree, export_graphviz
from sklearn.datasets import load_iris
%matplotlib inline
```

```python
import numpy as np
iris = load_iris()
```

```python
X,y = load_iris(return_X_y=True)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
print(iris)
```

# ID3

```python
dcf = DecisionTreeClassifier(criterion='entropy')
dcf.fit(X_train,y_train)
```

```python
import graphviz
```

```python
tree_data = export_graphviz(dcf, out_file=None, 
                            feature_names=iris.feature_names, 
                            class_names = iris.target_names,
                            filled=True, rounded=True, special_characters=True
                           )
```

```python
pred = dcf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_true=y_test, y_pred=pred, target_names=iris.target_names))
```

```python
graphviz.Source(tree_data)
```
